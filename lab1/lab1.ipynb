```python
import numpy as np
import pandas as pd
import pathlib
import os
from decimal import *

probabilitie =  pd.read_csv('./prob_08.csv', header=None)
table = pd.read_csv('./table_08.csv', header=None)


length= 20
```

```python
def unconditional_probability(cipher_text):
	df_table = table
	df_prob = probabilitie
	
	probability = 0.0

	for plain_text_index in range(length):
		for key_text_index in range(length):
			if (df_table[plain_text_index][key_text_index] == cipher_text):
				probability += df_prob[plain_text_index][0] * df_prob[key_text_index][1]
	
	return probability

df_cipher_texts_unconditional_prob =  np.array([unconditional_probability(i) for i in range(length)])

```

```python
def common_probability(cipher_text, plain_text):
	df_table = table
	df_prob = probabilitie
	
	probability = 0.0

	for key_index in range(length):
		if (df_table[plain_text][key_index] == cipher_text):
			probability += df_prob[plain_text][0] * df_prob[key_index][1]
	
	return probability

df_cipher_texts_common_prob =  np.array([[common_probability(i, j) for j in range(length)] for i in range(length)])

```

```python
def conditional_probability():
    return np.array([
			[df_cipher_texts_common_prob[i][j] / df_cipher_texts_unconditional_prob[i] for j in range(length)
		] for i in range(length)])

df_cipher_texts_conditional_prob =  conditional_probability()

```

```python
pd.DataFrame(df_cipher_texts_conditional_prob)
```

```python
def bayes_decision_function(cipher_text):
	return np.argmax(df_cipher_texts_conditional_prob[cipher_text])

df_bayes_decision_function =  [bayes_decision_function(i) for i in range(length)]

```

```python
pd.DataFrame({ 'C': range(length), 'M': df_bayes_decision_function }).transpose()
```

```python
def bayes_function_losses(cipher_text, plain_text):
	return 0 if df_bayes_decision_function[cipher_text] == plain_text else 1

def bayes_function_average_losses():
	return sum(
		[bayes_function_losses(i, j) * df_cipher_texts_common_prob[i][j] for j in range(length) for i in range(length)]
	)

df_bayes_function_average_losses =  bayes_function_average_losses()

```

```python

print(df_bayes_function_average_losses)
```

```python
def stochastic_decision_function():
	matrix = np.zeros((length, length))

	for i in range(length):
		current_probability_set = df_cipher_texts_conditional_prob[i]

		max_probability = max(current_probability_set)
		max_probability_entrance = list(current_probability_set).count(max(current_probability_set))

		for j in range(length):
			matrix[i][j] = 1 / max_probability_entrance if current_probability_set[j] == max_probability else 0

	return matrix

df_stochastic_decision_function = stochastic_decision_function()

```

```python
pd.DataFrame(df_stochastic_decision_function)
```

```python
def get_stochastic_function_losses(cipher_text, plain_text):
	return sum([df_stochastic_decision_function[cipher_text][i] for i in range(length) if i != plain_text])

def get_stochastic_function_average_losses():
	return sum(
		[get_stochastic_function_losses(i, j) * df_cipher_texts_common_prob[i][j] for j in range(length) for i in range(length)]
	)

df_stochastic_function_average_losses = get_stochastic_function_average_losses()

```
